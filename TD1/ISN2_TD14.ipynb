### Exercice 2 - Affichage de courbes

<u>Rappel :</u> Un guide de démarrage Matplotlib est disponible sur la page ISN - Ressources Communes sur Moodle : [version pur Python](https://moodle.insa-lyon.fr/pluginfile.php/421597/mod_resource/content/0/Matplotlib_pur_Python.pdf)  ou [version Numpy](https://moodle.insa-lyon.fr/pluginfile.php/421598/mod_resource/content/0/Matplolib_Python_Numpy.pdf) 

Le code ci-dessous, issu du TD2, utilise des listes python pour tracer une portion de la fonction $f(x) = x^2$.  Toutes les abscisses sont regroupées dans une liste (`x`) et les ordonnées correspondantes le sont dans une AUTRE liste (`fx`).

```python
# Affichage de la courbe $y = x^2$ 


x = [0, 1, 2, 3, 4, 5] 
fx = [0, 1, 4, 9, 16, 25] 

fig, ax = plt.subplots() 
ax.plot(x, fx, label="x=f(x)")     
ax.legend()

ax.set(xlabel='x', ylabel='f(x)', title='$y = x^2$ ')  
plt.show() 
```

---
On souhaite afficher la fonction $y=x^2$ de manière plus précise sur l'intervalle $[-10, 10]$.
Pour cela on propose d'utiliser la fonction `linspace(start, stop, nb_points)` de la bibliothèque `numpy`. Cette fonction renvoie un nd-array contenant `nb_points` réparties uniformément entre `start` et `stop`. Documentation de la fonction `linspace`: https://numpy.org/doc/stable/reference/generated/numpy.linspace.html


#### Q1. Utilisez la fonction `linspace` afin d'afficher la courbe de la fonction $y=x^2$ sur l'intervalle $[-10, 10]$ sur $100$ points.

```python
x = np.linspace(-10, 10, 100)
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y, label="$y = x^2$")
ax.set(xlabel='x', ylabel='y', title='Графік $x^2$ за допомогою Linspace')
ax.legend()
plt.show()
```

#### Q2. Faire une figure représentant la fonction $sin(x)$ et la fonction $sin(x+\frac{\pi}{2})$ sur l'intervalle $[0, 2\pi]$. Chaque courbe doit être d'une couleur différente et le graphique doit être légendé.

```python
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.sin(x + np.pi/2)

fig, ax = plt.subplots()
ax.plot(x, y1, color='blue', label='$sin(x)$')
ax.plot(x, y2, color='red', linestyle='--', label='$sin(x + \\pi/2)$')

ax.set(xlabel='x', ylabel='y', title='Тригонометричні функції')
ax.legend()
plt.show()
```

### Exercice A : Indexation, réduction et tranchage

#### <u> Diapos de support </u>: 14, 17-18, 20-21
Les données suivantes décrivent des populations de lièvres et de lynx (et de carottes) dans le nord du Canada sur 20 ans entre 1900 et 1920.

```python
annees = np.arange(1900, 1921)
lievres = np.array([30000, 47200, 70200, 77400, 36300, 20600, 18100, 21400, 22000, 25400, 27100, 40300, 57000, 76600, 52300, 19500, 11200, 7600, 14600, 16200, 24700])
lynx =  np.array([4000,  6100,  9800, 35200, 59400, 41700, 19000, 13000,  8300,  9100, 7400,  8000, 12300, 19500, 45700, 51100, 29700, 15800,  9700, 10100, 8600])
carottes = np.array([48300, 48200, 41500, 38200, 40600, 39800, 38600, 42300, 44500, 42100, 46000, 46800, 43800, 40900, 39400, 39000, 36700, 41800, 43300, 41300, 47300])
```

#### Q1. Visualisez ces 3 populations sur la période d'étude en traçant les 3 courbes sur le même graphique à l'aide de Matplotlib. N'oubliez pas le titre et les légendes.

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(annees, lievres, label='Lievres (Зайці)')
ax.plot(annees, lynx, label='Lynx (Рисі)')
ax.plot(annees, carottes, label='Carottes (Морква)')

ax.set(title='Еволюція популяцій (1900-1920)', xlabel='Роки', ylabel='Кількість')
ax.legend()
plt.show()
```

#### Q2. À l'aide de l'indexation, donnez le nombre d'individus de la population des lièvres à la première et à la dernière année de la période d'étude.

```python
print(f"Перший рік: {lievres[0]}")
print(f"Останній рік: {lievres[-1]}")
```

#### Q3. À l'aide des réductions, calculez le minimum, le maximum, la moyenne et l'écart-type de la population des lynx sur la période d'étude.

```python
print(f"Мінімум: {lynx.min()}")
print(f"Максимум: {lynx.max()}")
print(f"Середнє: {lynx.mean():.2f}")
print(f"Відхилення: {lynx.std():.2f}")
```

#### Q4. En quelle année la population des lynx est la plus faible ? 
Utilisez les réductions renvoyant des indices, c'est-à-dire les fonctions `np.arg*`.

```python
indice_min = np.argmin(lynx)
print(f"Найменша популяція рисей була у {annees[indice_min]} році.")
```

#### Q5. Quelles sont les deux valeurs de population des lynx les plus faibles ?
Utilisez la fonction [`np.sort`](https://numpy.org/doc/stable/reference/generated/numpy.sort.html) et le slicing.

```python
lynx_sorted = np.sort(lynx)
print(f"Дві найменші величини: {lynx_sorted[:2]}")
```

#### Q6. Quelles sont les deux années pour lesquelles la population des lynx est la plus faible ?

Utilisez la fonction [`np.argsort`](https://numpy.org/doc/stable/reference/generated/numpy.argsort.html) et le slicing.

```python
indices_sorted = np.argsort(lynx)
print(f"Два найгірші роки: {annees[indices_sorted[:2]]}")
```
